# 4. Answer the following questions with statistical evidence

### 4.a Do charges of people who smoke differ significantly from the people who don't?

```python
sns.distplot(ins[(ins.smoker == 0)]['charges'],color='b')
```

```python
sns.distplot(ins[(ins.smoker == 1)]['charges'],color='b')
```

#From the above charts, it is evident that charges are more for smokers as compared to non-smokers. For statistical significance,
we apply the 2 sample t-test

```python
non_smokers=ins[(ins.smoker == 0)]['charges']
smokers=ins[(ins.smoker == 1)]['charges']
```

```python
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro, wilcoxon
from statsmodels.stats.power import ttest_power
```

```python
t_statistic, p_value = ttest_ind(smokers, non_smokers)
print("two-sample t-test p-value=",t_statistic, p_value)
#Null hypothesis is that the two populations have the same mean. i.e, they are similar
```

```python
u, p_value = mannwhitneyu(smokers, non_smokers)
print ("two-sample wilcoxon-test p-value=", p_value)
```

```python
shapiro(smokers)
```

```python
shapiro(non_smokers)
```

In this case, Wilcoxon test also confirms to the same result as the 2-value t test. BUt regardless, we'd take the result of
Wilcoxon test since the distribution is not normal

Result: Since =, p-value is less than 0.05, we reject the null hypothesis and conclude 
that charges for smokers are significantly different from non-smokers

## 4.b.Does bmi of males differ significantly from that of females?

```python
bmi_males=ins[(ins.sex == 1)]['charges']
bmi_females=ins[(ins.sex == 0)]['charges']
```

```python
ins.head()
```

```python
#Two sample T-test
t_statistic, p_value = ttest_ind(bmi_males, bmi_females)
print(t_statistic, p_value)
#Null hypothesis is that the two populations have the same mean. i.e, they are similar
```

```python
# Result: Since =, p-value is less than 0.05, we reject the null hypothesis and conclude 
#that bmi for males are significantly different from females
```

```python
u, p_value = mannwhitneyu(bmi_males, bmi_females)
print ("two-sample wilcoxon-test p-value=", p_value)
```

```python
shapiro(bmi_males)
```

```python
shapiro(bmi_females)
```

Both distributions are not normal. Hence, we choose the result of the wilcoxon test over the 2 sample t-test and accept the
null hypothesis. Hence, bmi of males and females are not significantly different

## 4.d. Is the distribution of bmi across women with no children, one child and two children, the same ?


```python
child_0=ins[(ins.children == 0)][ins.sex==0]['bmi']
child_1=ins[(ins.children == 1)][ins.sex==0]['bmi']
child_2=ins[(ins.children == 2)][ins.sex==0]['bmi']
```

```python
shapiro(child_0)
```
