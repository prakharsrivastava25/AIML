```python
from collections import defaultdict
def get_top_n(predictions, n=5):
    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
top_n = get_top_n(predictions, n=5)
```

```python
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import Reader
```

```python
df = df_subset
# A reader is still needed but only the rating_scale param is requiered.
reader = Reader(rating_scale=(1, 5))

# The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(df[['user_id', 'prod_id', 'rating']], reader)
```

```python
trainset, testset = train_test_split(data, test_size=.30)
```

```python
# Using user_based collaborative filtering
from surprise import KNNWithMeans
algo_user_based = KNNWithMeans(k=100, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo_user_based.fit(trainset)
```

```python
# run the trained model against the testset
test_pred_user_based = algo_user_based.test(testset)
```

```python
# get RMSE
print("User-based collaborative filtering Model")
accuracy.rmse(test_pred_user_based, verbose=True)
```
